```python
import matplotlib.pyplot as plt

df_bra = df[df['Code'] == 'BRA']
df_usa = df[df['Code'] == 'USA']
df_chn = df[df['Code'] == 'CHN']

plt.figure(figsize=(9, 5))
plt.plot(df_bra['Year'], df_bra['Annual CO₂ emissions (per capita)'], label='Brasil', color='green')
plt.plot(df_usa['Year'], df_usa['Annual CO₂ emissions (per capita)'], label='Estados Unidos', color='blue')
plt.plot(df_chn['Year'], df_chn['Annual CO₂ emissions (per capita)'], label='China', color='red')

plt.title('Emissões de CO2: Brasil vs Estados Unidos vs China')
plt.xlabel('Ano')
plt.ylabel('Emissões anuais de CO₂ (per capita)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
